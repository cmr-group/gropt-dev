```python
import matplotlib.pyplot as plt
import numpy as np
from timeit import default_timer as timer

import gropt_dev as gropt
```

```python
gparams = gropt.GroptParams()
gparams.N = 120
gparams.dt = 10e-6
gparams.vec_init_simple(first_val = -.01, last_val = .02)

gparams.add_gmax(.08)
gparams.add_smax(200)
gparams.add_moment(0,0.0)
gparams.add_moment(1,14.0)


start_t = timer()
gparams.init()
gparams.solve()
out = gparams.get_out()
stop_t = timer()

print(f'Final Good: {gparams.final_good>0}   Solve time: {1000*(stop_t-start_t):.1f} ms')

plt.figure(figsize=(6,3))
tt_ms = np.arange(out.size)*gparams.dt*1e3
plt.plot(tt_ms, out)
plt.show()
```
