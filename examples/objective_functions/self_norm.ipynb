```python
import matplotlib.pyplot as plt
import numpy as np
from timeit import default_timer as timer

import gropt_dev as gropt
```

```python
# Minimize

gparams = gropt.GroptParams()
gparams.N = 101
gparams.dt = 10e-6
gparams.vec_init_simple()

gparams.add_gmax(.08)
gparams.add_smax(200)
gparams.add_moment(0,2.0)
gparams.add_moment(1,0.0)
gparams.add_moment(2,0.0)

I_weight = 1
gparams.add_obj_identity(I_weight)

start_t = timer()
gparams.init()
n_iter = 3000
gparams.solve(n_iter = n_iter, min_iter = n_iter-1000)
out = gparams.get_out()
stop_t = timer()

print(f'Weight = {I_weight:.1e}')
print(f'||g|| = {1000*np.linalg.norm(out):.1f}')
print(f'Solve time: {1000*(stop_t-start_t):.1f} ms')


plt.figure(figsize=(6,3))
tt_ms = np.arange(out.size)*gparams.dt*1e3
plt.plot(tt_ms, out)
plt.show()
```
